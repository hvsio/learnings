### Problem 1

##### Find the total number of crimes recorded in the CRIME table.

### Problem 2

##### List community areas with per capita income less than 11000.


```python
%sql SELECT * FROM CENSUS LIMIT 10
```

```python
%sql SELECT COMMUNITY_AREA_NAME FROM CENSUS WHERE per_capita_income < 11000
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%sql SELECT case_number FROM CRIME WHERE DESCRIPTION LIKE '%MINOR%'
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%sql SELECT case_number FROM CRIME WHERE PRIMARY_TYPE='KIDNAPPING'
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
%sql SELECT * FROM CRIME WHERE location_description LIKE '%SCHOOL%'
```

### Problem 6

##### List the average safety score for each type of school.


```python
%sql SELECT * FROM SCHOOLS LIMIT 10
```

```python
%sql SELECT AVG(SAFETY_SCORE), elementary__middle__or_high_school from SCHOOLS GROUP BY elementary__middle__or_high_school
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%sql SELECT * FROM CENSUS LIMIT 10
```

```python
%sql SELECT COMMUNITY_AREA_NAME, percent_households_below_poverty FROM CENSUS ORDER BY percent_households_below_poverty DESC LIMIT 5
```

### Problem 8

##### Which community area is most crime prone?


```python
%%sql
    SELECT COUNT(community_area_number) as CRIME_SCORE, community_area_number
    FROM CRIME
    GROUP BY community_area_number
    ORDER BY COUNT(community_area_number) DESC
    LIMIT 10
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql 
    SELECT COMMUNITY_AREA_NAME
    FROM CENSUS
    WHERE hardship_index = (SELECT MAX(hardship_index) FROM CENSUS)
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
    SELECT COMMUNITY_AREA_NAME AS FREQ
    FROM CENSUS
    WHERE COMMUNITY_AREA_NUMBER = (
        SELECT COMMUNITY_AREA_NUMBER, COUNT(*) FROM CRIME
        GROUP BY COMMUNITY_AREA_NAME
        )
```
